```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('medical_examination.csv')
df
```

```python
df['overweight'] = df['weight'] / (df['height']/100)
df
```

```python
df.loc[df['overweight'] <= 25, 'overweight'] = 0
```

```python
df.loc[df['overweight'] > 25, 'overweight'] = 1
```

```python
df['overweight'].unique()
```

```python
df
```

```python
df.loc[(df['cholesterol'] == 1) | (df['gluc'] == 1), ['cholesterol', 'gluc']] = 0
df.loc[(df['cholesterol'] > 1) | (df['gluc'] > 1), ['cholesterol', 'gluc']] = 1
```

```python
df
```

```python
df_cat = pd.melt(df, value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'], var_name='variable', value_name='value')
plt.figure(figsize=(10, 6))
sns.catplot(x='variable', hue='value', kind='count', data=df_cat, height=6, aspect=2)
```
